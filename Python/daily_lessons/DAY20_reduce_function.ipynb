## üîπ Python me reduce kaise use hota hai?

`reduce()` **built-in nahi** hota
‚Üí `functools` se import karna padta hai


```
from functools import reduce
```

### Basic syntax

```
reduce(function, iterable)
```

* `function` ‚Üí 2 inputs leta hai
* `iterable` ‚Üí list / tuple / etc.
* output ‚Üí **single value**

---

## üîπ Example 1: Sum of numbers


```
from functools import reduce

nums = [1, 2, 3, 4]

total = reduce(lambda a, b: a + b, nums)
print(total)
```

### Process inside (important):

```
a=1, b=2 ‚Üí 3
a=3, b=3 ‚Üí 6
a=6, b=4 ‚Üí 10
```

üëâ Final output = **10**

---

## üîπ Example 2: Maximum value


```
nums = [10, 5, 20, 8]

max_val = reduce(lambda a, b: a if a > b else b, nums)
print(max_val)
```

üëâ Output: **20**

---

## üîπ Example 3: ML-style (confidence aggregation)

Soch:

* Model ne probabilities di:


```
probs = [0.9, 0.8, 0.95]
```

* Overall confidence (minimum trust):

```
final_conf = reduce(lambda a, b: a if a < b else b, probs)
```

üëâ weakest link principle

## üîÅ map vs filter vs reduce (must remember)

| Function | Kya karta hai        | Output           |
| -------- | -------------------- | ---------------- |
| map      | Har element badalta  | List (same size) |
| filter   | Kuch elements chunta | List (smaller)   |
| reduce   | Sab ko mila deta     | **Single value** |
