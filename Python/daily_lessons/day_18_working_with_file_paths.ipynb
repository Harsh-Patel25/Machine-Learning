# üìÇ Python ‚Äì Working with File Paths

## ‚ú® Introduction

File handling tabhi useful hai jab tu file ka **path** sahi handle kare.

* **Relative Path** ‚Üí file ka location us program ke folder ke hisaab se.
* **Absolute Path** ‚Üí full address, jahan file computer me stored hai.

üëâ Example:

* Relative: `"data.txt"`
* Absolute: `"C:/Users/Shital/Documents/data.txt"` (Windows)
* Absolute: `"/home/shital/Documents/data.txt"` (Linux/Mac)

## üîë Module for Path Handling ‚Äì `os` & `pathlib`

Python me path ke liye do mast modules use hote hain:

* `os` (old but powerful)
* `pathlib` (modern aur easy)

## üõ† Example 1: Current Working Directory

```python
f = open("data.txt" , "w")
f.write("Hello World")
f.write("This is the my First Example of the File path")
f.close()
```

```python
import os

print(os.getcwd())
```

## üõ† Example 2: List Files in a Directory

```python
import os

files = os.listdir('.')
print(files)
```

## üõ† Example 3: Joining Paths Safely

```python
import os

folder = "D:/Machine Leaning A -  Zge-"
filename = "data.txt"
path = os.path.join(folder, filename)

print(path)  # Windows/Linux ke hisaab se correct format me path banega
```

üëâ `os.path.join()` use karna best hai, warna manually `/` dalne se galti ho sakti hai.

## ‚ö° Modern Way: Using `pathlib`

```python
from pathlib import Path

#Create Path object
path = Path('C:/Users/Shital/Documents/data.tx')

print("Name" , path.name) # file ka name
print("parents" , path.parents) # Parent folder
print("Suffix" , path.suffix) # Extesion of .txt
print("Exists")
```

## üìñ Example 4: Reading a File with Pathlib

```python
from pathlib import Path
path = Path("myfile.tx")

if  path.exists():
   content = path.read()
else:
  print("File not found!")
```

## üìÇ Example 5: Create Folder if Not Exist

```python
import os

folder = 'logs'

if not os.path.exists(folder):
  os.mkdir(folder)
  print("Folder is created" , folder)
else:
  print("Folder is alredy exists" , folder)

```
