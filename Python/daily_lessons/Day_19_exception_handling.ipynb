## ‚ú® Introduction
Errors aana normal hai jab program run hota hai. Agar unhe handle na kare to program crash ho jaata hai üö´.

## üîë Keywords in Exception Handling
1. **`try`** ‚Üí yaha wo code likho jo risky hai.
2. **`except`** ‚Üí agar error aaya to yaha handle hoga.
3. **`else`** ‚Üí agar error nahi aaya to ye chalega.
4. **`finally`** ‚Üí hamesha chalega (cleanup ke liye best).
5. **`raise`** ‚Üí custom error banane ke liye.


```python
f = open("sample_data.txt" , "w")
f.write("Bro Helllo my name is Har?sh Patel you know!\n")
f.write("Let's do some fun here with the execption hanling\n")
f.close()
```

```python
try:
  num = int(input("Enter the number:"))
  result = 10 / num
  print("Result" , result)
except:
  print("Somthing is wrong here")
```

üìí **Notes:**
- Agar user `0` daale ‚Üí ZeroDivisionError.
- Agar user `abc` daale ‚Üí ValueError.
- Yaha generic `except` use hua hai jo saare errors catch karega (production me avoid karo).


```python
try:
  num = int(input("Enter the number here:"))
  result = 10 / num
  print("Result" , result)
except ZeroDivisionError:
  print("Division by zero is not allowed here")
except ValueError:
  print("Please ente the valid number here")
```

üìí **Notes:**
- Specific exceptions handle karna best practice hai.
- Easy debugging aur custom messages ke liye useful.


```python
try:
  num = int(input("Enter the number:"))
  result = 10 / num
  print("Result" , result)
except ZeroDivisionError:
  print("Division by zero is not allowed here")
else:
  print("No error occured than Result is :" , result)

```

üìí **Notes:**
- `else` tabhi chalega jab try block successful ho aur koi error na aaye.


```python
try:
  with open("sample_data.txt","r") as f:
    for line in f:
      print(line, end='') # Add end='' to prevent extra newline
except FileNotFoundError:
  print("FIle not Found here")
finally:
  print("File operation is completed")
```

üìí **Notes:**
- `finally` hamesha chalega (error ho ya na ho).
- Mostly file closing, DB disconnect, cleanup ke liye use hota hai.


```python
def login(username , password):
  try:
    if username != "admin" or password != "1234":
      raise ValueError("Invalide username and password")
    print("Loggin is successfull!")
  except ValueError as e:
      print("Error:", e)

login("admin" , "1234")
login("admin" , "123")
```

üìí **Notes:**
- `raise` keyword custom error throw karne ke liye use hota hai.
- Useful jab tu apna validation implement karna chahta hai.


```python
try:
  num = [10 , 'a' , 0]
  for n in num:
    print(10 / n)
except (ZeroDivisionError , TypeError) as e:
  print("Error occured" , e)
```

üìí **Notes:**
- Ek `except` me multiple errors handle karne ke liye tuple use hota hai.
