## üîë File Operations Modes

Python me file kholne ke liye `open()` function use hota hai.

```python
open("filename", "mode")
```

**Modes**:

* `"r"` ‚Üí Read (default) ‚Äì file ko sirf read kar sakte ho.
* `"w"` ‚Üí Write ‚Äì file overwrite ho jaati hai (puranƒÅ data udh gaya üö´).
* `"a"` ‚Üí Append ‚Äì file ke end me add karta hai.
* `"x"` ‚Üí Create ‚Äì nayi file banata hai, agar file already exist kare to error.
* `"b"` ‚Üí Binary mode (images, videos ke liye).
* `"t"` ‚Üí Text mode (default).

## üõ† Basic Example 1: Writing to a File

````python
f = open("myfile.txt", "w")
f.write("Hello Python World!\n")
f.write("This is the my first file handling Example.")
f.close()
````

üëâ Isse ek file banegi `myfile.txt` jisme ye content hoga.

* `open("myfile.txt","w")` ‚Üí write mode me file open.
* `f.write(...)` ‚Üí file me content likhna.
* `f.close()` ‚Üí file close karna (important).

## üìñ Example 2: Reading from a File

````python
f = open("myfile.txt", "r")
content = f.read()
print(content)
f.close()
````

````python
f = open("myfile.txt", "r")
for line in f:
    print(line.strip())
f.close()
````

## ‚ûï Example 3: Appending to a File

````python
f = open("myfile.txt", "a")
f.write("\nNew line added after append.")
f.close()
````

## ‚ö° Pro Tip: With Statement (Best Practice)

````python
with open("myfile.txt" , "r") as f:
  print(f.read())
````

## üåç Real-World Example: Logging System

````python
import datetime

def log_event(event):
  with open("myfile.txt" , "a") as f:
    f.write(f"{datetime.datetime()} - {event}\n")

print("Loggin system is ready")
print(*"User logged in") # * use karo maja ayega  sare latter alag alg ho jayenge
````
