## 1ï¸âƒ£ Iterable vs Iterator (1-minute clarity)

* **Iterable**: jisme multiple values ho (list, tuple, string)
* **Iterator**: jo **ek-ek karke next value deta** hai


```
nums = [1, 2, 3]      # iterable
it = iter(nums)      # iterator

next(it)  # 1
next(it)  # 2
next(it)  # 3
```

ğŸ‘‰ Iterator **state yaad rakhta** hai (kahan tak pahuncha).

## 3ï¸âƒ£ Generator kya hota hai? (iterator ka smart version)

> **Generator = function jo `yield` use karta hai**
> Ye **lazy** hota hai: value **tabhi banata** hai jab maango.

### Simple generator


```
def count_up(n):
    for i in range(n):
        yield i
```

```
gen = count_up(3)
next(gen)  # 0
next(gen)  # 1
next(gen)  # 2
```

ğŸ‘‰ Function ruk-ruk ke value deta hai, **sab ek saath nahi**.

## 4ï¸âƒ£ Generator vs Normal Function

| Normal function     | Generator          |
| ------------------- | ------------------ |
| `return` once       | `yield` many times |
| Sab values ek saath | Ek-ek karke        |
| Memory heavy        | Memory efficient   |

## 5ï¸âƒ£ ML-style examples (real feel)

### ğŸ”¹ Large file processing


```
def read_lines(path):
    with open(path) as f:
        for line in f:
            yield line.strip()
```

```
for row in read_lines("big_data.csv"):
    process(row)
```

ğŸ‘‰ **Big data, small memory**.
